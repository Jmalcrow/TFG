```python
M=30
X_L=np.arange(0.4,0.6,0.001)
w1=1.
m=np.arange(1,M+1)
g=np.arange(0,2,0.01) #Here I have all the values of the coupling constant
MeanField=[]
MaxField=[]
#I will do everything for M=30 like in David's document
Delta_r_p1_M30=np.load("Delta_r_p1_M30.npy") #Here I have all the values of the renormalized qubit frequency
epsilon_r_p1_M30=np.load("epsilon_r_p1_M30.npy") #Here I have all the values of the renormalized bias
```

```python
def mean_field(coup,n,er,Dr,x_l):
    part_1=(coup*np.sin(n*np.pi/2)*np.sqrt(n)*er*np.sin(n*np.pi*x_l))/(np.sqrt(m*w1)*(1+0.003*(n-1))*np.sqrt(er*er + Dr*Dr))
    return (np.nansum(part_1)/(coup+0.0001))**2 #I put a safe value to no divide by zero
```

```python
#FOR THE MAX FIELD INSET
for i in range(np.size(g)):
    gaux=g[i]
    Dr_aux=Delta_r_p1_M30[i]
    er_aux=epsilon_r_p1_M30[i]
    x_l_max=0.5
    #now I just calculate the mean field
    field_max=mean_field(gaux,m,er_aux,Dr_aux,x_l_max)
    MaxField.append(field_max)
```
